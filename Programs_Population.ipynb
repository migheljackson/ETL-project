```python
prog_study_data=full_prog_data[['col_program_id',
                                'name',
                                'org_id',
                                'description',
                                'geographic_cluster_name',
                                'min_age',
                                'max_age',
                                'capacity',
                                'program_price',
                                'program_has_scholarships',
                                'program_provides_free_food',
                                'program_pays_participants',
                                'program_provides_transportation',
                                'category_name'
                               ]]
```
